```python
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
from datetime import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
```

```python
df_old = pd.read_csv('../CST Data Files/run_set_data_CST318.csv')
df_new = pd.read_csv('../CST Data Files/run_set_data_CST318m.csv')
```

```python
x1 = df_old['fermentation_age_hours']
y1 = df_old['air_ref.offgas1.argon_raw']

x2 = df_new['fermentation_age_hours']
y2 = df_new['air_ref.offgas1.argon_raw']

plt.plot(x1, y1, label='Old')
plt.plot(x2, y2, label='New')
plt.legend()
plt.title('Argon in')
plt.show()
```

```python
x1 = df_old['fermentation_age_hours']
y1 = df_old['offgas1.argon_raw']

x2 = df_new['fermentation_age_hours']
y2 = df_new['offgas1.argon_raw']

plt.plot(x1, y1, label='Old')
plt.plot(x2, y2, label='New')
plt.legend()
plt.title('Argon out')
plt.show()
```

```python
x1 = df_old['fermentation_age_hours']
y1 = df_old['cng_ref.offgas1.methane_raw']

x2 = df_new['fermentation_age_hours']
y2 = df_new['cng_ref.offgas1.methane_raw']

plt.plot(x1, y1, label='Old')
plt.plot(x2, y2, label='New')
plt.legend()
plt.title('CNG Methane in')
plt.show()
```

```python
x1 = df_old['fermentation_age_hours']
y1 = df_old['offgas1.methane_raw']

x2 = df_new['fermentation_age_hours']
y2 = df_new['offgas1.methane_raw']

plt.plot(x1, y1, label='Old')
plt.plot(x2, y2, label='New')
plt.legend()
plt.title('Methane out')
plt.show()
```

```python
x1 = df_old['fermentation_age_hours']
y1 = df_old['b1_ref.offgas1.methane_raw']

x2 = df_new['fermentation_age_hours']
y2 = df_new['b1_ref.offgas1.methane_raw']

plt.plot(x1, y1, label='Old')
plt.plot(x2, y2, label='New')
plt.legend()
plt.title('B1 Methane in')
plt.show()
```
